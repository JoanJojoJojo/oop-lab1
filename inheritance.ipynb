```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self, other=None):
        """Default greeting. Can optionally greet another Person."""
        if other:
            return f"{self.name} says hello to {other.name}."
        return f"{self.name} says hello."


class Joela(Person):
    def greet(self, other=None):
        # specialized greeting when Joela greets Eva
        if other and other.name.lower() == "eva":
            return f"{self.name}: 'Hi {other.name}! It's Joela â€” great to see you!'"
        # fallback to base behavior
        return super().greet(other)


class Eva(Person):
    def greet(self, other=None):
        # Eva responds warmly if someone greets her
        if other:
            return f"{self.name}: 'Hello {other.name}! How are you?'"
        return f"{self.name} waves hello."


# Example usage demonstrating inheritance and polymorphism
joela = Joela("Joela")
eva = Eva("Eva")
alex = Person("Alex")

# Joela greeting Eva (specialized behavior)
print(joela.greet(eva))   # Joela -> Eva

# Eva greeting Joela
print(eva.greet(joela))   # Eva -> Joela

# Polymorphism: treat different instances uniformly
people = [joela, eva, alex]
for p in people:
    # call the same method on different subclasses/instances
    print(p.greet())
```
